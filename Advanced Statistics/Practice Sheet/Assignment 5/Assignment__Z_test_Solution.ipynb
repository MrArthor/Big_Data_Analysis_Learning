*Q1-* A university claims that its graduates earn an average starting salary of $60,000 per year.

You take a sample of 100 graduates and find the average salary to be $58,000

with a known population standard deviation of $10,000.

Test the university's claim at a 0.05 significance level.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

'''
Null Hypothesis (𝐻0): The average salary of graduates is $60,000 (𝜇=60,000).
Alternate Hypothesis (𝐻𝑎): The average salary of graduates is not $60,000 (𝜇≠60,000).

'''

# Given data
sample_mean = 58000
population_mean = 60000
population_std = 10000
sample_size = 100
alpha = 0.05

# Calculate Z-score
z_Test = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Determine critical Z-values for a two-tailed test
z_critical = stats.norm.ppf(1 - alpha / 2)

# Display results
result = {
    "Sample Mean": sample_mean,
    "Population Mean": population_mean,
    "Population Standard Deviation": population_std,
    "Sample Size": sample_size,
    "Z-Score": round(z_Test, 2),
    "Z-Critical": round(z_critical, 2),
    "Reject Null Hypothesis?": abs(z_Test) > z_critical
}

# Visualization
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Standard Normal Distribution")
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color="red", alpha=0.3, label="Rejection Region")
plt.axvline(z_Test, color="blue", linestyle="--", label=f"Z-Test = {z_Test:.2f}")
plt.axvline(-z_critical, color="black", linestyle="--", label=f"Z-Critical = ±{z_critical:.2f}")
plt.axvline(z_critical, color="black", linestyle="--")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.title("Z-Test for University Salary Claim")
plt.legend()
plt.show()
```

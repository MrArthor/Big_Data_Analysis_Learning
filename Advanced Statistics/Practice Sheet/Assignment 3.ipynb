## Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
mean_weight = 500  # cm
std_dev_weight = 10  # cm

# Generate a range of weights around the mean
weights = np.linspace(480, 535, 500)

# Calculate PDF and CDF values
pdf_values = norm.pdf(weights, mean_weight, std_dev_weight)
cdf_values = norm.cdf(weights, mean_weight, std_dev_weight)

# Plotting the PDF and CDF
plt.figure(figsize=(14, 6))

# Plot the PDF
plt.subplot(1, 2, 1)
plt.plot(weights, pdf_values, label="PDF", color="blue")
plt.axvline(490, color="orange", linestyle="--", label="weight = 180 cm")
plt.axvline(495, color="green", linestyle="--", label="weight = 165 cm")
plt.axvline(520, color="red", linestyle="--", label="weight = 185 cm")
plt.title("Probability Density Function (PDF) of weights")
plt.xlabel("weight (cm)")
plt.ylabel("Probability Density")
plt.legend()

# Plot the CDF
plt.subplot(1, 2, 2)
plt.plot(weights, cdf_values, label="CDF", color="purple")
plt.axvline(490, color="orange", linestyle="--", label="weight = 180 cm")
plt.axvline(495, color="green", linestyle="--", label="weight = 165 cm")
plt.axvline(520, color="red", linestyle="--", label="weight = 185 cm")
plt.title("Cumulative Distribution Function (CDF) of weights")
plt.xlabel("weight (cm)")
plt.ylabel("Cumulative Probability")
plt.legend()

plt.tight_layout()
plt.show()
```

## Question 3

```python
from scipy.stats import poisson

# Given data
calls_per_hour = 10  # average rate

# (a) Probability of receiving exactly 8 calls in an hour
prob_exactly_8_calls = poisson.pmf(8, calls_per_hour)
print("P(exactly 8 calls):", prob_exactly_8_calls)

# (b) Probability of receiving 15 or fewer calls in an hour
prob_15_or_fewer_calls = poisson.cdf(15, calls_per_hour)
print("P(15 or fewer calls):", prob_15_or_fewer_calls)

# (c) Probability of receiving more than 20 calls in an hour
prob_more_than_20_calls = 1 - poisson.cdf(20, calls_per_hour)
print("P(more than 20 calls):", prob_more_than_20_calls)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given Poisson mean (lambda)
lambda_calls = 10

# Normal approximation parameters
mean = lambda_calls
std_dev = np.sqrt(lambda_calls)

# Generate a range of call counts around the mean
x_values = np.linspace(0, 30, 500)
pdf_values = norm.pdf(x_values, mean, std_dev)

# Plotting the normal approximation curve
plt.figure(figsize=(12, 6))
plt.plot(x_values, pdf_values, label="Normal Approximation", color="blue")

# Shade area for exactly 8 calls (between 7.5 and 8.5)
plt.fill_between(x_values, 0, pdf_values, where=(x_values >= 7.5) & (x_values <= 8.5),
                 color="orange", alpha=0.5, label="P(exactly 8 calls)")

# Shade area for 15 or fewer calls
plt.fill_between(x_values, 0, pdf_values, where=(x_values <= 15),
                 color="green", alpha=0.3, label="P(15 or fewer calls)")

# Shade area for more than 20 calls
plt.fill_between(x_values, 0, pdf_values, where=(x_values > 20),
                 color="red", alpha=0.3, label="P(more than 20 calls)")

# Labels and title
plt.title("Normal Approximation for Call Center Calls per Hour")
plt.xlabel("Number of Calls")
plt.ylabel("Probability Density")
plt.legend()

plt.show()
```

## Qustion 4

```python
from scipy.stats import poisson

# Given Poisson mean (lambda) for accidents per day
lambda_accidents = 3

# (a) Probability of observing exactly 5 accidents
prob_exactly_5 = poisson.pmf(5, lambda_accidents)
print("P(exactly 5 accidents):", prob_exactly_5)

# (b) Probability of observing no more than 2 accidents
prob_no_more_than_2 = poisson.cdf(2, lambda_accidents)
print("P(no more than 2 accidents):", prob_no_more_than_2)

# (c) Distribution Explanation:
print("\nSince accidents are relatively rare events occurring over a fixed period (days),"
      " a Poisson distribution is more suitable than a Normal distribution for modeling the data.")
import numpy as np
import matplotlib.pyplot as plt

# Generate possible accident counts for Poisson distribution
x_values = np.arange(0, 11)  # Range of possible accident counts
pmf_values = poisson.pmf(x_values, lambda_accidents)  # PMF values
cdf_values = poisson.cdf(x_values, lambda_accidents)  # CDF values

# Plotting PMF and CDF
plt.figure(figsize=(14, 6))

# (a) Poisson PMF for exact counts of accidents
plt.subplot(1, 2, 1)
plt.stem(x_values, pmf_values, basefmt=" ")
plt.title("Poisson PMF for Accident Counts")
plt.xlabel("Number of Accidents")
plt.ylabel("Probability")
plt.fill_between([5 - 0.5, 5 + 0.5], 0, poisson.pmf(5, lambda_accidents), color="orange", alpha=0.5, label="P(exactly 5 accidents)")
plt.legend()

# (b) Poisson CDF for cumulative probability of accidents
plt.subplot(1, 2, 2)
plt.plot(x_values, cdf_values, color="purple")
plt.title("Poisson CDF for Accident Counts")
plt.xlabel("Number of Accidents")
plt.ylabel("Cumulative Probability")
plt.fill_between(x_values, cdf_values, where=(x_values <= 2), color="green", alpha=0.3, label="P(no more than 2 accidents)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import poisson
# Given data
arrival_rate = 5  # students per minute

# (a) Probability of exactly 3 students arriving in the next minute
prob_exactly_3_arrivals = poisson.pmf(3, arrival_rate)
print("P(exactly 3 students arrive):", prob_exactly_3_arrivals)

# (b) Probability of at least 10 students arriving within the next 2 minutes
# Adjust rate for 2 minutes
prob_10_or_more_arrivals_in_2_min = 1 - poisson.cdf(9, arrival_rate * 2)
print("P(at least 10 students in 2 minutes):", prob_10_or_more_arrivals_in_2_min)

# (c) Probability that waiting time for the next student exceeds 20 seconds (normal distribution)
mean_wait_time = 12  # seconds
std_dev_wait_time = 5  # seconds
prob_wait_exceeds_20 = 1 - norm.cdf(20, mean_wait_time, std_dev_wait_time)
print("P(waiting time > 20 seconds):", prob_wait_exceeds_20)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm

# Given data
arrival_rate_per_min = 5  # students per minute
arrival_rate_per_2min = 2 * arrival_rate_per_min  # Adjusted for 2 minutes
mean_wait_time = 12  # seconds
std_dev_wait_time = 5  # seconds

# Generate possible student counts for Poisson distribution
x_values_poisson = np.arange(0, 21)

# Calculate Poisson PMF for exactly 3 students in one minute
pmf_values = poisson.pmf(x_values_poisson, arrival_rate_per_min)

# Calculate Poisson CDF for 10 or more students in two minutes
x_values_cdf = np.arange(0, 31)
cdf_values_2min = poisson.cdf(x_values_cdf, arrival_rate_per_2min)

# Plot Poisson PMF
plt.figure(figsize=(15, 6))

# (a) Poisson PMF for exactly 3 students arriving in the next minute
plt.subplot(1, 3, 1)
plt.stem(x_values_poisson, pmf_values, basefmt=" ")
plt.title("Poisson PMF for Student Arrivals (1 min)")
plt.xlabel("Number of Students")
plt.ylabel("Probability")
plt.fill_between([3 - 0.5, 3 + 0.5], 0, poisson.pmf(3, arrival_rate_per_min), color="orange", alpha=0.5, label="P(exactly 3 students)")
plt.legend()

# (b) Poisson CDF for at least 10 students in 2 minutes
plt.subplot(1, 3, 2)
plt.plot(x_values_cdf, poisson.cdf(x_values_cdf, arrival_rate_per_2min), color="purple")
plt.fill_between(x_values_cdf, poisson.cdf(x_values_cdf, arrival_rate_per_2min), where=(x_values_cdf >= 10), color="green", alpha=0.3)
plt.title("Poisson CDF for Student Arrivals (2 min)")
plt.xlabel("Number of Students")
plt.ylabel("Cumulative Probability")

# (c) Normal distribution for waiting times between arrivals
wait_times = np.linspace(0, 30, 100)
pdf_wait_times = norm.pdf(wait_times, mean_wait_time, std_dev_wait_time)

plt.subplot(1, 3, 3)
plt.plot(wait_times, pdf_wait_times, color="blue", label="Waiting Time Distribution")
plt.fill_between(wait_times, pdf_wait_times, where=(wait_times >= 20), color="red", alpha=0.5, label="P(waiting time > 20 seconds)")
plt.title("Normal Distribution of Waiting Times")
plt.xlabel("Waiting Time (seconds)")
plt.ylabel("Probability Density")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Given data
mean_purchase = 75  # dollars
std_dev_purchase = 15  # dollars

# (a) Probability that a randomly selected customer spends more than $100
prob_spends_more_than_100 = 1 - norm.cdf(100, mean_purchase, std_dev_purchase)
print("P(spends > $100):", prob_spends_more_than_100)

# (b) Probability that a customer spends between $60 and $90
prob_spends_between_60_and_90 = norm.cdf(90, mean_purchase, std_dev_purchase) - norm.cdf(60, mean_purchase, std_dev_purchase)
print("P($60 < spends < $90):", prob_spends_between_60_and_90)

# (c) Probability of receiving more than 25 purchases in an hour
purchases_per_hour = 20  # average rate
prob_more_than_25_purchases = 1 - poisson.cdf(25, purchases_per_hour)
print("P(more than 25 purchases):", prob_more_than_25_purchases)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
mean_purchase = 75  # dollars
std_dev_purchase = 15  # dollars

# Generate a range of purchase amounts around the mean
purchase_amounts = np.linspace(30, 120, 500)
pdf_values = norm.pdf(purchase_amounts, mean_purchase, std_dev_purchase)

# Plotting the Normal PDF
plt.figure(figsize=(12, 6))
plt.plot(purchase_amounts, pdf_values, label="Purchase Amount Distribution", color="blue")

# Shade area for spending more than $100
plt.fill_between(purchase_amounts, 0, pdf_values, where=(purchase_amounts > 100),
                 color="orange", alpha=0.5, label="P(spending > $100)")

# Shade area for spending between $60 and $90
plt.fill_between(purchase_amounts, 0, pdf_values, where=(purchase_amounts >= 60) & (purchase_amounts <= 90),
                 color="green", alpha=0.3, label="P($60 \leq spending \leq $90)")

# Labels and title
plt.title("Normal Distribution of Customer Purchase Amounts")
plt.xlabel("Purchase Amount ($)")
plt.ylabel("Probability Density")
plt.legend()

plt.show()
```

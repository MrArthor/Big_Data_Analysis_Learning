1. **Titanic Dataset**
Description: Information about Titanic passengers, including demographics, survival status, and ticket class.
- Link: Titanic Dataset on Kaggle
- Columns: Age, Fare, Class, Sex, Embarked, and Survival status.
- Use Cases: Sampling based on survival class, analyzing central tendency of Age and Fare.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_df = pd.read_csv(url)

# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Separate the dataset based on survival status
survived_df = titanic_df[titanic_df['Survived'] == 1]
not_survived_df = titanic_df[titanic_df['Survived'] == 0]

# Sample 5 rows from each class
sample_survived = survived_df.sample(5, random_state=1)
sample_not_survived = not_survived_df.sample(5, random_state=1)

# Display samples
print("Sampled Survived:")
print(sample_survived)
print("\nSampled Not Survived:")
print(sample_not_survived)
```

```python
# Plot the survival distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Survived')
plt.title('Survival Distribution on Titanic')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate central tendency measures for 'Age' and 'Fare'
age_mean = titanic_df['Age'].mean()
age_median = titanic_df['Age'].median()
age_mode = titanic_df['Age'].mode()[0]

fare_mean = titanic_df['Fare'].mean()
fare_median = titanic_df['Fare'].median()
fare_mode = titanic_df['Fare'].mode()[0]

# Print central tendency
print(f"Age - Mean: {age_mean}, Median: {age_median}, Mode: {age_mode}")
print(f"Fare - Mean: {fare_mean}, Median: {fare_median}, Mode: {fare_mode}")

```

```python
# Plot Age distribution with central tendency indicators
plt.figure(figsize=(12, 5))

# Age plot
plt.subplot(1, 2, 1)
sns.histplot(titanic_df['Age'].dropna(), bins=30, kde=True)
plt.axvline(age_mean, color='r', linestyle='--', label=f'Mean: {age_mean:.2f}')
plt.axvline(age_median, color='g', linestyle='-', label=f'Median: {age_median:.2f}')
plt.axvline(age_mode, color='b', linestyle='-', label=f'Mode: {age_mode}')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()

# Fare plot
plt.subplot(1, 2, 2)
sns.histplot(titanic_df['Fare'], bins=30, kde=True)
plt.axvline(fare_mean, color='r', linestyle='--', label=f'Mean: {fare_mean:.2f}')
plt.axvline(fare_median, color='g', linestyle='-', label=f'Median: {fare_median:.2f}')
plt.axvline(fare_mode, color='b', linestyle='-', label=f'Mode: {fare_mode}')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

**Wine Quality Dataset**
mDescription: Chemical composition and quality ratings for red and white wines.
Link: Wine Quality Dataset on UCI Machine Learning Repository
Columns: Fixed acidity, volatile acidity, citric acid, residual sugar, alcohol, pH, quality.
Use Cases: Sampling wine quality levels, measuring central tendencies of acidity, sugar, and alcohol content

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Wine Quality dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_df = pd.read_csv(url, sep=";")

# Display the first few rows of the dataset
wine_df.head()


```

```python
# Check unique quality levels
quality_levels = wine_df['quality'].unique()
print("Quality Levels:", quality_levels)

# Sample 5 rows from each quality level
sampled_qualities = wine_df.groupby('quality').apply(lambda x: x.sample(5, random_state=1))

# Display the sampled data
print("Sampled Data by Quality Level:")
print(sampled_qualities)
```

```python
# Plot the distribution of wine quality
plt.figure(figsize=(8, 6))
sns.countplot(data=wine_df, x='quality', palette="viridis")
plt.title('Distribution of Wine Quality Levels')
plt.xlabel('Wine Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate central tendency measures for 'fixed acidity', 'residual sugar', and 'alcohol'
fixed_acidity_mean = wine_df['fixed acidity'].mean()
fixed_acidity_median = wine_df['fixed acidity'].median()
fixed_acidity_mode = wine_df['fixed acidity'].mode()[0]

residual_sugar_mean = wine_df['residual sugar'].mean()
residual_sugar_median = wine_df['residual sugar'].median()
residual_sugar_mode = wine_df['residual sugar'].mode()[0]

alcohol_mean = wine_df['alcohol'].mean()
alcohol_median = wine_df['alcohol'].median()
alcohol_mode = wine_df['alcohol'].mode()[0]

# Print central tendency results
print(f"Fixed Acidity - Mean: {fixed_acidity_mean}, Median: {fixed_acidity_median}, Mode: {fixed_acidity_mode}")
print(f"Residual Sugar - Mean: {residual_sugar_mean}, Median: {residual_sugar_median}, Mode: {residual_sugar_mode}")
print(f"Alcohol - Mean: {alcohol_mean}, Median: {alcohol_median}, Mode: {alcohol_mode}")
```

```python
# Plot the distributions of fixed acidity, residual sugar, and alcohol content with central tendencies
plt.figure(figsize=(15, 5))

# Fixed Acidity plot
plt.subplot(1, 3, 1)
sns.histplot(wine_df['fixed acidity'], bins=30, kde=True, color="purple")
plt.axvline(fixed_acidity_mean, color='r', linestyle='--', label=f'Mean: {fixed_acidity_mean:.2f}')
plt.axvline(fixed_acidity_median, color='g', linestyle='-', label=f'Median: {fixed_acidity_median:.2f}')
plt.axvline(fixed_acidity_mode, color='b', linestyle='-', label=f'Mode: {fixed_acidity_mode}')
plt.title('Fixed Acidity Distribution')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.legend()

# Residual Sugar plot
plt.subplot(1, 3, 2)
sns.histplot(wine_df['residual sugar'], bins=30, kde=True, color="orange")
plt.axvline(residual_sugar_mean, color='r', linestyle='--', label=f'Mean: {residual_sugar_mean:.2f}')
plt.axvline(residual_sugar_median, color='g', linestyle='-', label=f'Median: {residual_sugar_median:.2f}')
plt.axvline(residual_sugar_mode, color='b', linestyle='-', label=f'Mode: {residual_sugar_mode}')
plt.title('Residual Sugar Distribution')
plt.xlabel('Residual Sugar')
plt.ylabel('Frequency')
plt.legend()

# Alcohol plot
plt.subplot(1, 3, 3)
sns.histplot(wine_df['alcohol'], bins=30, kde=True, color="blue")
plt.axvline(alcohol_mean, color='r', linestyle='--', label=f'Mean: {alcohol_mean:.2f}')
plt.axvline(alcohol_median, color='g', linestyle='-', label=f'Median: {alcohol_median:.2f}')
plt.axvline(alcohol_mode, color='b', linestyle='-', label=f'Mode: {alcohol_mode}')
plt.title('Alcohol Content Distribution')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

Adult Income Dataset
Description: Data on income classification based on census data, with attributes like age, education, occupation, and income category.
Link: Adult Income Dataset on UCI Machine Learning Repository
Columns: Age, Work class, Education, Occupation, Income level (>50K or <=50K).
Use Cases: Stratified sampling based on income or education, central tendencies of age and work hours.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Adult Income dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
columns = ['Age', 'Workclass', 'fnlwgt', 'Education', 'Education-Num', 'Marital-Status',
           'Occupation', 'Relationship', 'Race', 'Sex', 'Capital-Gain',
           'Capital-Loss', 'Hours-per-week', 'Native-Country', 'Income']

# Read the dataset
adult_income_df = pd.read_csv(url, header=None, names=columns, na_values=' ?', skipinitialspace=True)

# Display the first few rows of the dataset
adult_income_df.head()
```

```python
# Check for missing values
print(adult_income_df.isnull().sum())

# Display the data types and basic statistics
print(adult_income_df.info())
print(adult_income_df.describe())
```

```python
# Stratified sampling based on Income level
stratified_sample = adult_income_df.groupby('Income').apply(lambda x: x.sample(5, random_state=1)).reset_index(drop=True)

# Display the stratified sample
print("Stratified Sample based on Income:")
print(stratified_sample)
```

```python
# Calculate central tendency measures for 'Age' and 'Hours-per-week'
age_mean = adult_income_df['Age'].mean()
age_median = adult_income_df['Age'].median()
age_mode = adult_income_df['Age'].mode()[0]

hours_mean = adult_income_df['Hours-per-week'].mean()
hours_median = adult_income_df['Hours-per-week'].median()
hours_mode = adult_income_df['Hours-per-week'].mode()[0]

# Print central tendency results
print(f"Age - Mean: {age_mean:.2f}, Median: {age_median:.2f}, Mode: {age_mode:.2f}")
print(f"Hours-per-week - Mean: {hours_mean:.2f}, Median: {hours_median:.2f}, Mode: {hours_mode:.2f}")
```

```python
# Plot the distributions of Age and Hours-per-week
plt.figure(figsize=(12, 5))

# Age distribution
plt.subplot(1, 2, 1)
sns.histplot(adult_income_df['Age'], bins=20, kde=True, color='blue')
plt.axvline(age_mean, color='r', linestyle='--', label=f'Mean: {age_mean:.2f}')
plt.axvline(age_median, color='g', linestyle='-', label=f'Median: {age_median:.2f}')
plt.axvline(age_mode, color='b', linestyle='-', label=f'Mode: {age_mode:.2f}')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()

# Hours-per-week distribution
plt.subplot(1, 2, 2)
sns.histplot(adult_income_df['Hours-per-week'], bins=20, kde=True, color='orange')
plt.axvline(hours_mean, color='r', linestyle='--', label=f'Mean: {hours_mean:.2f}')
plt.axvline(hours_median, color='g', linestyle='-', label=f'Median: {hours_median:.2f}')
plt.axvline(hours_mode, color='b', linestyle='-', label=f'Mode: {hours_mode:.2f}')
plt.title('Hours-per-week Distribution')
plt.xlabel('Hours-per-week')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

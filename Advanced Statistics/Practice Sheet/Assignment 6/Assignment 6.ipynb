```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chisquare

# Observed frequencies of ice cream flavors
flavors = ['Chocolate', 'Vanilla', 'Strawberry', 'Mint', 'Mango']
observed_frequencies = [90, 60, 40, 30, 20]

# Expected frequencies (assuming equal preference for all flavors)
total_observed = sum(observed_frequencies)
num_flavors = len(observed_frequencies)
expected_frequencies = [total_observed / num_flavors] * num_flavors

# Perform the Chi-Square Goodness of Fit test
chi2, p = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Print the results
print("Chi-Square Statistic:", chi2)
print("p-value:", p)

# Hypothesis Testing Decision
alpha = 0.05
if p <= alpha:
    print(f"Reject the null hypothesis (p-value = {p} ≤ α = {alpha}). Preferences are not equally distributed across flavors.")
else:
    print(f"Fail to reject the null hypothesis (p-value = {p} > α = {alpha}). Preferences are equally distributed across flavors.")

# Visualization 1: Observed vs Expected Frequencies (Bar Chart)
df = pd.DataFrame({
    'Flavor': flavors,
    'Observed': observed_frequencies,
    'Expected': expected_frequencies
})

df.plot(x='Flavor', kind='bar', stacked=False)
plt.title('Observed vs Expected Frequencies of Ice Cream Flavors')
plt.xlabel('Flavor')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Observed vs Expected Frequencies (Heatmap)
plt.figure(figsize=(8, 5))
sns.heatmap(df.set_index('Flavor').T, annot=True, cmap="YlGnBu", fmt=".0f")
plt.title('Observed and Expected Frequencies (Heatmap)')
plt.xlabel('Flavor')
plt.ylabel('Frequency Type')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Data for Gender and Pet Ownership
data = {
    'Owns Pet': [70, 60],
    'Does Not Own Pet': [50, 80]
}
index = ['Male', 'Female']

# Convert to a DataFrame
df = pd.DataFrame(data, index=index)

# Perform the Chi-Square Test of Independence
chi2, p, dof, expected = chi2_contingency(df)

# Print the results
print("Chi-Square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:\n", pd.DataFrame(expected, index=index, columns=df.columns))

# Hypothesis Testing Decision
alpha = 0.05
if p <= alpha:
    print(f"Reject the null hypothesis (p-value = {p} ≤ α = {alpha}). There is a significant association between gender and pet ownership.")
else:
    print(f"Fail to reject the null hypothesis (p-value = {p} > α = {alpha}). There is no significant association between gender and pet ownership.")

# Visualization 1: Observed Frequencies (Bar Chart)
df.plot(kind='bar', stacked=True)
plt.title('Observed Frequencies of Gender vs. Pet Ownership')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.legend(title='Pet Ownership')
plt.show()

# Visualization 2: Expected Frequencies (Heatmap)
expected_df = pd.DataFrame(expected, index=index, columns=df.columns)
plt.figure(figsize=(8, 5))
sns.heatmap(expected_df, annot=True, fmt=".2f", cmap="Blues")
plt.title('Expected Frequencies (Heatmap)')
plt.xlabel('Pet Ownership')
plt.ylabel('Gender')
plt.show()
```

```python
# Observed frequencies for each car color
observed_frequencies = np.array([120, 90, 70, 50, 30])
colors = ['White', 'Black', 'Silver', 'Blue', 'Red']

# Expected frequencies assuming equal distribution
expected_frequencies = np.array([sum(observed_frequencies) / len(observed_frequencies)] * len(observed_frequencies))

# Perform chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
print("Expected frequencies:", expected_frequencies)

# Plot heatmap of observed data
plt.figure(figsize=(8, 6))
sns.heatmap(observed_frequencies.reshape(1, -1), annot=True, fmt="d", cmap="YlGnBu", xticklabels=colors, yticklabels=["Observed"])
plt.xlabel("Car Color")
plt.title("Car Color Preference (Observed Counts)")
plt.show()

# Bar plot for observed vs expected frequencies
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(colors))
width = 0.35

ax.bar(x - width / 2, observed_frequencies, width, label='Observed')
ax.bar(x + width / 2, expected_frequencies, width, label='Expected', alpha=0.5)

ax.set_xlabel('Car Color')
ax.set_ylabel('Frequency')
ax.set_title('Observed vs Expected Car Color Preferences')
ax.set_xticks(x)
ax.set_xticklabels(colors)
ax.legend()

plt.show()
```

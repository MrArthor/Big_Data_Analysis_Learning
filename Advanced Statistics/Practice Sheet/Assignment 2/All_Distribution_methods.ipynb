**To perform various distributions and visualize them using Python, you can follow the steps outlined below. This code will include implementations for the following distributions:**

- Normal/Gaussian Distribution
- Standard Normal Distribution
- Z-Score Calculation
- Log-Normal Distribution
- Binomial Distribution
- Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set the style for seaborn
sns.set(style="whitegrid")

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution data

# 1. Normal/Gaussian Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, color='skyblue', stat='density')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 2. Standard Normal Distribution
mean = 0
std_dev = 1
x = np.linspace(-4, 4, 100)
standard_normal = stats.norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, standard_normal, color='orange')
plt.title('Standard Normal Distribution (Mean=0, SD=1)')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='red', lw=0.5, ls='--')
plt.show()

# 3. Z-Score Calculation
z_scores = (data - np.mean(data)) / np.std(data)
plt.figure(figsize=(10, 6))
sns.histplot(z_scores, bins=30, kde=True, color='lightgreen', stat='density')
plt.title('Z-Score Distribution')
plt.xlabel('Z-Score')
plt.ylabel('Density')
plt.show()

# 4. Log-Normal Distribution
log_normal_data = np.random.lognormal(mean=0, sigma=0.954, size=1000)
plt.figure(figsize=(10, 6))
sns.histplot(log_normal_data, bins=30, kde=True, color='purple', stat='density')
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 5. Binomial Distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
binomial_data = np.random.binomial(n, p, 1000)
plt.figure(figsize=(10, 6))
sns.histplot(binomial_data, bins=np.arange(0, n + 1) - 0.5, stat='density', color='cyan', kde=False)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.xticks(np.arange(0, n + 1))
plt.show()

# 6. Bernoulli Distribution
bernoulli_data = np.random.binomial(1, p, 1000)
plt.figure(figsize=(10, 6))
sns.histplot(bernoulli_data, bins=[-0.5, 0.5, 1.5], stat='density', color='pink', kde=False)
plt.title('Bernoulli Distribution (p=0.5)')
plt.xlabel('Outcome (0 or 1)')
plt.ylabel('Density')
plt.xticks([0, 1])
plt.show()

```

**Explanation of the Code:**
- Normal/Gaussian Distribution: Creates a histogram of normally distributed data with a density curve overlay.

- Standard Normal Distribution: Plots the probability density function (PDF) for the standard normal distribution (mean = 0, standard deviation = 1).

- Z-Score Calculation: Computes the z-scores of the normal data and visualizes their distribution.

- Log-Normal Distribution: Generates and plots a log-normal distribution.

- Binomial Distribution: Samples data from a binomial distribution and visualizes it.

- Bernoulli Distribution: Samples from a Bernoulli distribution (a special case of the binomial distribution with n=1) and visualizes the outcomes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Create a synthetic dataset
np.random.seed(0)
data_size = 1000
data = {
    'Value': np.random.normal(loc=50, scale=10, size=data_size),  # Normal distribution data
    'Binary': np.random.binomial(n=1, p=0.5, size=data_size)  # Bernoulli distribution
}
df = pd.DataFrame(data)

# Display the first few rows of the dataset
print(df.head())

# 1. Normal/Gaussian Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Value'], bins=30, kde=True, color='skyblue', stat='density')
plt.title('Normal Distribution of Values')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 2. Standard Normal Distribution
mean = df['Value'].mean()
std_dev = df['Value'].std()
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
standard_normal = stats.norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, standard_normal, color='orange')
plt.title('Standard Normal Distribution (Mean={:.2f}, SD={:.2f})'.format(mean, std_dev))
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(mean, color='red', lw=0.5, ls='--')
plt.show()

# 3. Z-Score Calculation
z_scores = (df['Value'] - mean) / std_dev
plt.figure(figsize=(10, 6))
sns.histplot(z_scores, bins=30, kde=True, color='lightgreen', stat='density')
plt.title('Z-Score Distribution')
plt.xlabel('Z-Score')
plt.ylabel('Density')
plt.show()

# 4. Log-Normal Distribution
log_normal_data = np.random.lognormal(mean=np.log(mean), sigma=std_dev/mean, size=data_size)
plt.figure(figsize=(10, 6))
sns.histplot(log_normal_data, bins=30, kde=True, color='purple', stat='density')
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 5. Binomial Distribution
n = 1  # Number of trials for Bernoulli
p = df['Binary'].mean()  # Probability of success
binomial_data = np.random.binomial(n, p, data_size)
plt.figure(figsize=(10, 6))
sns.histplot(binomial_data, bins=[-0.5, 0.5, 1.5], stat='density', color='cyan', kde=False)
plt.title('Binomial Distribution (n=1, p={:.2f})'.format(p))
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.xticks([0, 1])
plt.show()

# 6. Bernoulli Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Binary'], bins=[-0.5, 0.5, 1.5], stat='density', color='pink', kde=False)
plt.title('Bernoulli Distribution (p={:.2f})'.format(p))
plt.xlabel('Outcome (0 or 1)')
plt.ylabel('Density')
plt.xticks([0, 1])
plt.show()
```

**Explanation of the Code:**

- Synthetic Dataset Creation:

    - A dataset (df) is created with two columns:
    - Value: Contains 1000 values drawn from a normal distribution (mean = 50, standard deviation = 10).
    - Binary: Contains 1000 values from a Bernoulli distribution with a probability of success of 0.5.
- Normal Distribution Visualization:

    - A histogram and KDE plot are generated for the Value column.
- Standard Normal Distribution:

  - The PDF of the standard normal distribution is plotted using the mean and standard deviation of the Value column.
- Z-Score Calculation:

  - Z-scores are calculated for the Value data and visualized.
- Log-Normal Distribution:

- A log-normal distribution is visualized using randomly generated log-normal data based on the mean and standard deviation.
- Binomial Distribution:

  - A binomial distribution is created and visualized based on the Binary column.
- Bernoulli Distribution:

  - The distribution of the binary outcomes is visualized.

```python
## Applying Gaussian (Normal) Distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a Gaussian-distributed dataset
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)

# Visualize the Gaussian distribution
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color="blue", bins=30)
plt.title("Gaussian (Normal) Distribution")
plt.xlabel("Data values")
plt.ylabel("Frequency")
plt.show()
```

```python
##  Applying Log Transformation

# Applying log transformation (shift data by a constant to avoid log(0))
log_data = np.log(data - min(data) + 1)

# Visualize the log-transformed data
plt.figure(figsize=(10, 6))
sns.histplot(log_data, kde=True, color="green", bins=30)
plt.title("Log-Transformed Data")
plt.xlabel("Log-transformed values")
plt.ylabel("Frequency")
plt.show()
```

```python
## Applying Central Limit Theorem (CLT)

# Applying Central Limit Theorem
sample_means = [np.mean(np.random.choice(log_data, size=30)) for _ in range(1000)]

# Visualize the distribution of sample means
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, kde=True, color="purple", bins=30)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample means")
plt.ylabel("Frequency")
plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate log-distributed data
np.random.seed(0)
data = np.random.lognormal(mean=2.0, sigma=0.5, size=1000)

# Visualize the log-distributed data
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color="orange", bins=30)
plt.title("Log-Distributed Data")
plt.xlabel("Log values")
plt.ylabel("Frequency")
plt.show()

# Step 2: Apply Box-Cox transformation (requires positive values)
data_boxcox, _ = stats.boxcox(data)

# Visualize the transformed (approximately Gaussian) data
plt.figure(figsize=(10, 6))
sns.histplot(data_boxcox, kde=True, color="blue", bins=30)
plt.title("Data after Box-Cox Transformation (Approx. Gaussian)")
plt.xlabel("Box-Cox transformed values")
plt.ylabel("Frequency")
plt.show()
```

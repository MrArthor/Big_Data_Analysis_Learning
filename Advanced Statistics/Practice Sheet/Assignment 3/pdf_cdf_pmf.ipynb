PDF (**Probability Density Function**), CDF (**Cumulative Distribution Function**), and PMF (**Probability Mass Function**)
- Python, let's use scipy and matplotlib for calculations and visualizations. -
- PDF and CDF are generally used for continuous distributions, while PMF is used for discrete distributions.

**Poisson distribution**

We use the Poisson distribution to model the probability of a given number of events occurring within a fixed interval when the events happen independently and at a constant average rate.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

# Define the range for our x-axis
x = np.linspace(-4, 4, 10)

# Parameters for Normal Distribution
mu, sigma = 0, 1

# Step 1: Calculate CDF for Normal Distribution
cdf_normal = norm.cdf(x, mu, sigma)

# Step 2: Calculate PDF from CDF for Normal Distribution using np.gradient (numerical derivative)
pdf_from_cdf_normal = np.gradient(cdf_normal, x)

# Printing step-by-step values for PDF from CDF
print("Normal Distribution (CDF to PDF):")
for i in range(10):  # Show first 5 calculations as an example
    print(f"x[{i}] = {x[i]:.3f}, CDF[{i}] = {cdf_normal[i]:.5f}, PDF[{i}] (from CDF) = {pdf_from_cdf_normal[i]:.5f}")
```

```python
# Define values for discrete distribution (Poisson) for PMF and CDF
x_discrete = np.arange(0, 10)
lambda_poisson = 10

# Step 3: Calculate CDF for Poisson Distribution
cdf_poisson = poisson.cdf(x_discrete, lambda_poisson)

# Step 4: Calculate PMF from CDF for Poisson Distribution by finding differences
pmf_from_cdf_poisson = np.diff(cdf_poisson, prepend=0)

# Printing step-by-step values for PMF from CDF
print("\nPoisson Distribution (CDF to PMF):")
for i in range(len(x_discrete)):  # Show all calculations for discrete points
    print(f"x_discrete[{i}] = {x_discrete[i]}, CDF[{i}] = {cdf_poisson[i]:.5f}, PMF[{i}] (from CDF) = {pmf_from_cdf_poisson[i]:.5f}")
```

```python
# Plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 12))

# Plot PDF from CDF (Normal)
ax[0].plot(x, pdf_from_cdf_normal, color='blue', label="PDF derived from CDF (Normal)")
ax[0].set_title("PDF Derived from CDF - Normal Distribution")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Density")
ax[0].legend()

# Plot CDF (Normal)
ax[1].plot(x, cdf_normal, color='green', label="CDF (Normal)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Normal Distribution")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Cumulative Probability")
ax[1].legend()

# Plot PMF derived from CDF (Poisson)
ax[2].stem(x_discrete, pmf_from_cdf_poisson, basefmt=" ", linefmt='red', markerfmt="ro", label="PMF derived from CDF (Poisson)")
ax[2].set_title("PMF Derived from CDF - Poisson Distribution")
ax[2].set_xlabel("X (Discrete)")
ax[2].set_ylabel("Probability")
ax[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

# Define the range for our x-axis
x = np.linspace(-4, 4, 50)

# PDF for a normal (Gaussian) distribution
mu, sigma = 0, 1
pdf = norm.pdf(x, mu, sigma)

# CDF for a normal distribution
cdf = norm.cdf(x, mu, sigma)

# Define values for discrete distribution (Poisson) for PMF
x_discrete = np.arange(0, 50)
lambda_poisson = 5
pmf = poisson.pmf(x_discrete, lambda_poisson)

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 12))

# Plot PDF
ax[0].plot(x, pdf, color='blue', label="PDF (Normal)")
ax[0].set_title("Probability Density Function (PDF) - Normal Distribution")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Density")
ax[0].legend()

# Plot CDF
ax[1].plot(x, cdf, color='green', label="CDF (Normal)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Normal Distribution")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Cumulative Probability")
ax[1].legend()

# Plot PMF
ax[2].stem(x_discrete, pmf, basefmt=" ", linefmt='red', markerfmt="ro", label="PMF (Poisson)")
ax[2].set_title("Probability Mass Function (PMF) - Poisson Distribution")
ax[2].set_xlabel("X (Discrete)")
ax[2].set_ylabel("Probability")
ax[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

# Define the range for our x-axis
x = np.linspace(-4, 4, 1000)

# Parameters for Normal Distribution
mu, sigma = 0, 1

# CDF and PDF for Normal Distribution
cdf_normal = norm.cdf(x, mu, sigma)
pdf_from_cdf_normal = np.gradient(cdf_normal, x)  # Derivative of CDF to get PDF

# Define values for discrete distribution (Poisson) for PMF and CDF
x_discrete = np.arange(0, 15)
lambda_poisson = 5

# CDF and PMF for Poisson Distribution
cdf_poisson = poisson.cdf(x_discrete, lambda_poisson)
pmf_from_cdf_poisson = np.diff(cdf_poisson, prepend=0)  # Differences in CDF for PMF

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 12))

# Plot PDF from CDF (Normal)
ax[0].plot(x, pdf_from_cdf_normal, color='blue', label="PDF derived from CDF (Normal)")
ax[0].set_title("PDF Derived from CDF - Normal Distribution")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Density")
ax[0].legend()

# Plot CDF (Normal)
ax[1].plot(x, cdf_normal, color='green', label="CDF (Normal)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Normal Distribution")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Cumulative Probability")
ax[1].legend()

# Plot PMF derived from CDF (Poisson)
ax[2].stem(x_discrete, pmf_from_cdf_poisson, basefmt=" ", linefmt='red', markerfmt="ro", label="PMF derived from CDF (Poisson)")
ax[2].set_title("PMF Derived from CDF - Poisson Distribution")
ax[2].set_xlabel("X (Discrete)")
ax[2].set_ylabel("Probability")
ax[2].legend()

plt.tight_layout()
plt.show()
```
